```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open("time_vs_loc.p", "rb") as pf:
    result_dictionary = pickle.load(pf)
```

```python
for number_loc in result_dictionary.keys():
    relavant= result_dictionary[number_loc]
    pathlen = []
    for time in result_dictionary[number_loc].keys():
        r = relavant[time][0]
        pathlen.append(r)
    plt.plot(relavant.keys(), pathlen, label=str(number_loc))
plt.legend()
plt.ylim(0, 90)
```

```python
for number_loc in result_dictionary.keys():
    relavant= result_dictionary[number_loc]
    rewards = []
    for time in result_dictionary[number_loc].keys():
        r = relavant[time][1]
        rewards.append(r)
    plt.plot(relavant.keys(), rewards, label=str(number_loc))
plt.legend()
plt.ylim(0, 50)
```

```python
for number_loc in result_dictionary.keys():
    relavant= result_dictionary[number_loc]
    best_time = []
    for time in result_dictionary[number_loc].keys():
        r = relavant[time][2]
        best_time.append(r)
    plt.plot(relavant.keys(), best_time, label=str(number_loc))
plt.legend()
plt.ylim(0, 6500)
```
